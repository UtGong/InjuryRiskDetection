```python
import opensim as osim
import matplotlib.pyplot as plt
import os

# Example path to a .trc file (adjust as needed)
motion_file = "/home/ubuntu/injury_detection/SMPL2AddBiomechanics/output/01/trials/01_01_poses.trc"

if not os.path.isfile(motion_file):
    raise FileNotFoundError(f"Could not find TRC file at {motion_file}")

# Load the motion file using OpenSim's Storage class
storage = osim.Storage(motion_file)

# --- MANUAL INFO PRINTING --- #
num_rows = storage.getSize()  # Number of time frames/rows
first_time = storage.getFirstTime()
last_time = storage.getLastTime()

# Retrieve column labels (returns an ArrayStr)
column_labels_osim = storage.getColumnLabels()
column_labels = [column_labels_osim.get(i) for i in range(column_labels_osim.getSize())]

print("=== Storage File Info ===")
print(f"File: {motion_file}")
print(f"Number of rows (time frames): {num_rows}")
print(f"Time range: {first_time} to {last_time} seconds")
print(f"Column labels: {column_labels}")
print("=========================")

# Set the target column from the available columns. For example, choose "gu_x"
target_label = "gu_x"

time_data = []
target_data = []

if target_label in column_labels:
    label_index = column_labels.index(target_label)
    for i in range(num_rows):
        state_vec = storage.getStateVector(i)
        t = state_vec.getTime()
        time_data.append(t)
        
        # Extract the data vector for the current row
        data_vec = state_vec.getData()
        # Build a list from the OpenSim data vector
        row_data = [data_vec.get(j) for j in range(data_vec.getSize())]
        target_data.append(row_data[label_index])
    
    # Plot the selected column over time
    plt.figure(figsize=(8, 4))
    plt.plot(time_data, target_data, label=target_label)
    plt.xlabel("Time (s)")
    plt.ylabel("Data Value")
    plt.title(f"{target_label} Over Time")
    plt.legend()
    plt.grid(True)
    plt.show()
else:
    print(f"Column '{target_label}' not found in motion file columns.")
```

```python
# 1. Load model - suppress geometry warnings
osim.Logger.setLevel(osim.Logger.Level_Warn)  # Reduce warning verbosity
model = osim.Model("/home/ubuntu/injury_detection/SMPL2AddBiomechanics/models/bsm/bsm.osim")
model.initSystem()
print("model loaded successfully")
```

```python
# 2. Setup Inverse Dynamics Tool
id_tool = osim.InverseDynamicsTool()
print("Inverse DynamicsTool created successfully")

# Version-agnostic configuration
id_tool.setModel(model)
id_tool.setName("InverseDynamicsAnalysis")  # Required in some versions
print("Set model and name successfully")

# Set input files (using absolute paths)
mot_file = os.path.abspath("/home/ubuntu/injury_detection/SMPL2AddBiomechanics/models/bsm/sample_motion/01/01_01_ik.mot")
id_tool.setCoordinatesFileName(mot_file)
print("Set coordinates file successfully")

# Set output options
id_tool.setResultsDir("ID_Results")
id_tool.setOutputGenForceFileName("inverse_dynamics.sto")
id_tool.setLowpassCutoffFrequency(6.0)  # Must be float
print("Set output options successfully")
```

```python
import opensim as osim
import subprocess
import time

# 1. First try the normal way (with timeout)
try:
    print("Attempting standard execution...")
    start_time = time.time()
    id_tool.run()  # This might still hang
    
    # Timeout check
    if time.time() - start_time > 120:  # 2 minute timeout
        raise TimeoutError("Execution timed out")
        
except Exception as e:
    print(f"Standard method failed: {str(e)}")
    
    # 2. Fallback to command line execution
    print("Switching to command line execution...")
    setup_file = "id_setup_autogen.xml"
    id_tool.printToXML(setup_file)
    
    # Run with explicit path to opensim-cmd
    cmd = f"opensim-cmd run-tool {setup_file}"
    process = subprocess.Popen(cmd, shell=True, 
                              stdout=subprocess.PIPE, 
                              stderr=subprocess.PIPE)
    
    # Wait with timeout
    try:
        stdout, stderr = process.communicate(timeout=180)  # 3 minute timeout
        print(stdout.decode())
        if stderr:
            print("Errors:", stderr.decode())
    except subprocess.TimeoutExpired:
        process.kill()
        print("Command line execution timed out")
```

```python
# 3. Run the analysis
try:
    print("Running inverse dynamics...")
    id_tool.run()
    print("Success! Results saved to ID_Results/inverse_dynamics.sto")
except Exception as e:
    print(f"Error running ID: {str(e)}")
    print("Trying alternative approach...")
    
    # Alternative: Write setup file and execute via command line
    id_tool.printToXML("id_setup.xml")
    os.system("opensim-cmd run-tool id_setup.xml")
```
